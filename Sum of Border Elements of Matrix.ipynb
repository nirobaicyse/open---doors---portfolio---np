```python
# Problem 9 - Sum of Border Elements of Matrix
n = int(input())
mat = [list(map(int, input().split())) for _ in range(n)]
s = 0
for i in range(n):
    for j in range(n):
        if i == 0 or j == 0 or i == n-1 or j == n-1:
            s += mat[i][j]
print(s)


```
