```python
# Project 3: Iris Flower Classification
# Name: Md Alim Hossain | HSC Math: 186/200 (93%)
# Open Doors 2025-26 | Computer and Data Science

# Step 1: Import libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Step 2: Load Iris dataset
print("Loading Iris dataset...")
iris = load_iris()
X = iris.data      # Features: sepal length, width, petal length, width
y = iris.target    # Target: 0 = Setosa, 1 = Versicolor, 2 = Virginica

# Step 3: Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train KNN model
print("Training KNN model...")
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

# Step 5: Predict and show accuracy
pred = model.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(f"\nAccuracy: {accuracy * 100:.1f}% → 100.0% means perfect classification!")

# Step 6: Predict for a new flower
new_flower = [[5.1, 3.5, 1.4, 0.2]]  # Example: Setosa
flower_name = iris.target_names[model.predict(new_flower)[0]]
print(f"Prediction: This flower is → {flower_name.upper()}")
```
