```python
# Project 2: Salary Prediction using Linear Regression
# Name: Md Alim Hossain | HSC Math: 186/200 (93%)
# Open Doors 2025-26 | Computer and Data Science

# Step 1: Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Step 2: Create sample data (Age vs Salary)
ages = np.array([[25], [30], [35], [40], [45]])    # Input (X)
salaries = np.array([50000, 60000, 70000, 80000, 90000])  # Output (y)

# Step 3: Train the model
model = LinearRegression()
model.fit(ages, salaries)

# Step 4: Predict salary for age 38
predicted_salary = model.predict([[38]])[0]
print(f"Predicted Salary for Age 38: {predicted_salary:.0f} (BDT/Taka)")

# Step 5: Plot the data and regression line
plt.figure(figsize=(8, 5))
plt.scatter(ages, salaries, color='blue', label='Actual Data')
plt.plot(ages, model.predict(ages), color='red', label='Prediction Line')
plt.title("Salary Prediction based on Age")
plt.xlabel("Age (years)")
plt.ylabel("Salary (BDT)")
plt.legend()
plt.grid(True)
plt.show()
```
